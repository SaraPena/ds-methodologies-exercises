3. Cake Recipes

      a. Read the `cake_recipes.csv` data. This data set contains cake tastiness scores for combination of different recipes, oven rack positions, and oven temperatures.

```python
df = pd.read_csv('./untidy_data/cake_recipes.csv')
df
```

b. Tidy the data as necessary.

First we'll need to melt so that we can combine the temperature columns together, then we'll split up the recipe:position column into two, and reshape the data

```python
df = df.melt(id_vars = 'recipe:position', var_name = 'temperature')
df.head(6)
```

We'll need to split up the `recipe:position` column into two. We can do this with the Series.str.split method, which iwll give us back a dataframe when we can then concatenate onto our original data, then remove the old column.

```python
recipe_position_df = df['recipe:position'].str.split(':', expand = True)
```

```python
recipe_position_df.head(10)
```

```python
recipe_position_df.columns = ['recipe', 'position']
```

```python
recipe_position_df.head(10)
```

```python
df2 = pd.concat([df, recipe_position_df], axis = 1)
df2
```

```python
df2.drop(columns = 'recipe:position', inplace = True)
df2
```

c. Which recipe on average, is the best?

```python
recipes = list(df2.recipe.unique())
recipes
```

```python
best_score = 0
for r in recipes:
    score = df2[df2.recipe == r].value.mean()
    if score > best_score:
        best_score = score
        best_avg_score = {}
        best_avg_score[r] = best_score
best_avg_score
    
```

```python
df2.groupby('recipe').value.mean()
```

d. Which oven temperature, on average, produces the best results? 

```python
temperatures = list(df2.temperature.unique())
temperatures
```

```python
best_score = 0
for t in temperatures:
    score = df2[df2.temperature == t].value.mean()
    if score > best_score:
        best_score = score
        best_temp = t
        #best_avg_score = {}
        #best_avg_score[t] == best_score
best_score, best_temp
```

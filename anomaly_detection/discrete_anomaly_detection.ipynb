```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.rc('font', size = 14)
import seaborn as sns
```

```python
colnames=['ip', 'timestamp', 'request_method', 'status', 'size',
          'destination', 'request_agent']
df_orig = pd.read_csv('http://python.zach.lol/access.log',          
                 engine='python',
                 header=None,
                 index_col=False,
                 names=colnames,
                 sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
                 na_values='"-"',
                 usecols=[0, 3, 4, 5, 6, 7, 8]
)

new = pd.DataFrame([["95.31.18.119", "[21/Apr/2019:10:02:41+0000]", 
                     "GET /api/v1/items/HTTP/1.1", 200, 1153005, np.nan, 
                     "python-requests/2.21.0"],
                    ["95.31.16.121", "[17/Apr/2019:19:36:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 301, 1005, np.nan, 
                     "python-requests/2.21.0"],
                    ["97.105.15.120", "[18/Apr/2019:19:42:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 301, 2560, np.nan, 
                     "python-requests/2.21.0"],
                    ["97.105.19.58", "[19/Apr/2019:19:42:41+0000]", 
                     "GET /api/v1/sales?page=79/HTTP/1.1", 200, 2056327, np.nan, 
                     "python-requests/2.21.0"]], columns=colnames)

df = df_orig.append(new).reset_index().drop(columns = 'index')
```

```python
df
```

```python
df.timestamp = df.timestamp.str.replace(r'(\[|\])', '', regex=True)
df.timestamp= pd.to_datetime(df.timestamp.str.replace(':', ' ', 1))
df.sort_values(by='timestamp',inplace=True)
df = df.set_index('timestamp')
df
```

```python
for col in ['request_method', 'request_agent', 'destination']:
    df[col] = df[col].str.replace('"', '')

df['request_method'] = df.request_method.str.replace(r'\?page=[0-9]+', '', regex=True)
```

```python
df
```

```python
method_df = pd.DataFrame(df.request_method.value_counts(dropna=False)).reset_index().\
                rename(index=str, columns={'index': 'request_method', 'request_method': 'method_count'})
```

```python
method_df
```

```python
method_df2 = pd.DataFrame(df.request_method.value_counts(dropna=False)/df.request_method.count()).reset_index().\
                rename(index=str, columns={'index': 'request_method', 'request_method': 'method_proba'})
method_df2.sort_values(by='method_proba')
```

```python
method_df = method_df.merge(method_df2)
method_df
```

```python
method_df[method_df.method_proba < .01]
```

```python
print(len(method_df))

print(method_df.tail(10))

plt.figure(figsize=(12, 4))
splot = sns.barplot(data=method_df, x = 'request_method', y = 'method_count', ci = None)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', xytext = (0, 10), 
                   textcoords = 'offset points'
                   )
    plt.xticks(rotation='vertical')
```

```python
train = df['2019-04-16 19:34:42':'2019-04-17 12:55:14'][['ip','request_method','status','size','destination','request_agent']]
train
```

```python
method_df = pd.DataFrame(train.request_method.value_counts(dropna=False)/train.request_method.count()).reset_index().\
                rename(index=str, columns={'index': 'request_method', 'request_method': 'method_proba'})
method_df
```

```python
df = df.reset_index().merge(method_df, on=['request_method'], how='left').fillna(value=0).set_index('timestamp')
```

```python
df
```

```python
df.method_proba.value_counts()
```

```python
method_probs = train.groupby('request_method').size().div(len(df))
method_probs
```

```python
status_given_method = pd.DataFrame(train.groupby(['request_method', 'status']).\
                               size().div(len(train)).\
                               div(method_probs, 
                                   axis=0, 
                                   level='request_method').\
                               reset_index().\
                               rename(index=str, 
                                      columns={0: 'proba_status_given_method'})
                              )
status_given_method
```

```python
method_status_count = pd.DataFrame(train.groupby(['request_method', 'status'])['ip'].\
                                count().reset_index().rename(columns={'ip':'count'}))
method_status_count
```

```python
method_status = status_given_method.merge(method_status_count)
method_status
```

```python
df = df.reset_index().merge(method_status, on=['request_method', 'status'], how='left').fillna(value=0).set_index('timestamp')
```
